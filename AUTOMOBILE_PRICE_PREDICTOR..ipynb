## Now making machine learning model with the new dataset after performing backward elimination

```python
dataset = pd.read_csv('IMPORT85.CSV')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# from sklearn.compose import ColumnTransformer
# from sklearn.preprocessing import OneHotEncoder
# ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0,1,2,3,4,5,8])], remainder='passthrough')
# X = np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
X[:,0:6]=encoder.fit_transform(X[:,0:6])
```

```python
X[:,0:6]
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
X[:,8:9]=encoder.fit_transform(X[:,8:9])
```

```python
X[:,8:9]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Training the model on the training set

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=5,random_state=0)
regressor.fit(X_train,y_train)
```

## Predicting the new results

```python
y_pred=regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

## Evaluating the performance of the machine learning model

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
